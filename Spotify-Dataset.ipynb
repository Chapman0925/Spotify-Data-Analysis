## 1. What are the Top 10 songs in the spotify platform with 90+ Popularity?

```python
top10song=tracks.query('popularity>90').sort_values(by='popularity',ascending=False).head(10)
top10song
```

```python
plt.figure(figsize=(15,9))
sns.barplot(data=top10song, x='name', y= 'popularity')
plt.ylabel('Popularity', size = 20)
plt.xlabel("Name of Song", size = 20)
plt.title('Top 10 Most Popular Spotify Songs', size=20)
plt.xticks(rotation=90)
plt.show()

print('1.Peaches,2.drivers license,3.Astronaut the Ocean')
```

## 2. Data Preprocessing

```python
tracks.set_index('release_date', inplace = True)
```

```python
tracks.index=pd.to_datetime(tracks.index)
```

```python
tracks['duration']=tracks['duration_ms'].apply(lambda x:round(x/1000))
```

```python
tracks.drop(columns='duration_ms', axis=1,inplace=True)
```

```python
tracks.head()
```

```python
tracks.duration.head(5)
```

## 3. Finding the correlation between each features

```python
corr_tracks = tracks.drop(columns=['mode','explicit','key'],axis=1)
```

```python
corr_tracks= corr_tracks.corr(method='pearson')
```

```python
plt.figure(figsize=(18,9))
hm=sns.heatmap(corr_tracks,annot=True, fmt='0.2f', vmin=-1, vmax=1,center=0, cmap='flare',linewidth=1,linecolor='black')
hm.set_title('Correlation Heatmap between each Variables')
hm.set_xticklabels(hm.get_xticklabels(), rotation = 30)
hm.set_yticklabels(hm.get_yticklabels(), rotation = 30)
plt.show()
```

```python
sample=tracks.sample(int(len(tracks)/250))
len(sample)
```

```python
plt.figure(figsize=(15,6))
sns.regplot(data=sample, x='loudness',y='energy',scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.title('Loudness vs Energy', size=17)
plt.show()

print('Loudness & Energy has postivie correlation with each other and it has correlation of 0.76!')
```

## 4. How many songs are produced during each year?

```python
tracks.insert(2,'Year',tracks.index.get_level_values('release_date'))
```

```python
tracks['Year']=tracks['Year'].dt.year
```

```python
year=tracks['Year']
```

```python
plt.figure(figsize=(19,9))
sns.displot(year,discrete=True,aspect=2,height=8,kind='hist')
plt.title('Total # of songs of each year from 1922-2021', size=17)
plt.show()
```

## 5. What are the average duration of songs for each year?

```python
dur=tracks['duration']
plt.figure(figsize=(19,10))
sns.barplot(x=year,y=dur,errwidth=False)
plt.title('Average Duration of songs in each year', size= 18)
plt.ylabel('Duration', size = 17)
plt.xlabel("Year", size = 17)
plt.xticks(rotation=90)
plt.show()
```

## 6. What are the duration of different type of songs?
